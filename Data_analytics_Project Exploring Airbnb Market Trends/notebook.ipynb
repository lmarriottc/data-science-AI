Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# We've loaded your first package for you! You can add as many cells as you need.
import numpy as np
import os

os.listdir('data/'
          )
```

```python
# Loading data
import pandas as pd

price_location = pd.read_csv('data/airbnb_price.csv')
room_type = pd.read_excel('data/airbnb_room_type.xlsx')
last_review = pd.read_csv('data/airbnb_last_review.tsv', sep='\t', header=0, date_parser=True)

display(price_location.head())
display(room_type.head())
display(last_review.head())
```

```python
# What are the dates of the earliest and most recent reviews? 

last_review['last_review'] = pd.to_datetime(last_review['last_review'])
last_review.info()
```

```python
dates_ordered = last_review['last_review'].sort_values(ascending=True)
earliest_date = dates_ordered.iloc[0]
latest_date = dates_ordered.iloc[-1]
print('early:',earliest_date, 'latest:', latest_date)
```

```python
# How many of the listings are private rooms? 

room_type['room_type'] = room_type['room_type'].str.lower()
private_rooms = room_type[room_type['room_type'] == 'private room']
count_private_rooms = private_rooms['room_type'].value_counts()
#count_private_rooms = int(count_private_rooms)
```

```python
(room_type['room_type'].str.lower()).value_counts()
```

```python
# What is the average listing price?

price_location['price_int'] = price_location['price'].str.extract('(\d+)')
price_location['price_int'] = price_location['price_int'].astype(int)
mean_price = price_location['price_int'].mean().round(2)
print(mean_price)
```

```python
# Dataframe with results

review_dates = pd.DataFrame({
    'first_reviewed': earliest_date,
    'last_reviewed': latest_date,
    'nb_private_rooms': count_private_rooms,
    'avg_price': mean_price
})

#review_dates = review_dates.drop('index', axis=1)

display(review_dates)
```
