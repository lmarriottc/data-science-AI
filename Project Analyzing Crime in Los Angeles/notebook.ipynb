```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
crimes['HOUR OCC'] = crimes['TIME OCC'].str[:2].astype(int)
```

```python
# Which hour has the highest frequency of crimes?

sns.countplot(data = crimes, x='HOUR OCC')

plt.show()
```

```python
peak_crime_hour = 12
```

```python
#Which area has the larget frequency of night crimes (between 10pm and 3h59am)?

night_crimes = crimes[(crimes['HOUR OCC'] < 4) | (crimes['HOUR OCC'] >= 22)]
```

```python
print(night_crimes['HOUR OCC'].value_counts().sort_values(ascending=False))
```

```python
sns.countplot(data=night_crimes, x='AREA NAME')
plt.xticks(rotation=90)
plt.show()

#area_crimes = night_crimes.groupby('HOUR OCC')['AREA NAME'].count()
```

```python
crimes['Vict Age'] = crimes['Vict Age'].apply(lambda x: 0 if x < 0 else x)
```

```python
crimes['Vict Age'].describe()
```

```python
# Number of crimes commited against victims by age group(<18, 18-25, 26-34, 35-44, 45-54, 55-64, 65+)

age_bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
age_labels = ['<18', '18-25', '26-34', '35-44', '45-54', '55-64', '65+']

# Cut the data into bins
crimes['Age Bracket'] = pd.cut(crimes['Vict Age'], bins=age_bins, labels=age_labels)
```

```python
victim_ages = crimes['Age Bracket'].value_counts()
print(victim_ages)
```
